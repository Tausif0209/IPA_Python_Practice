```python
def oddeven(num):
  odd=0
  even=0
  num=abs(num)
  num=str(num)
  for d in num:
    if int(d)%2==0:
      even+=1
    else:odd+=1
  return odd,even
num=int(input("enter a number"))
odd,even=oddeven(num)
print(f"number:-{num}\nodd digit:-{odd}\neven digit:-{even}")
```

```python
import logging
logging.basicConfig(
    filename="vowels.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def vowelCount(s:str)->int:
  """
  Count the number of vowels in a given string.
    Parameters:
        s (str): Input string
    Returns:
        int: Total number of vowels in the string.
  """
  vowels=0
  for ch in s:
    if ch.lower() in "aeiou":
     vowels+=1
  return vowels  
s=input("Enter a String")
logging.info(f"String:-{s} vowels count:-{vowelCount(s)}")
```

```python
import logging
logging.basicConfig(
  filename="frequency.log",
  level=logging.INFO,
  format="%(asctime)s-%(levelname)s-%(message)s")
def fre(key:str,s:str)->int:
  """
  Count the frequency of given character  in a given string.
    Parameters:
        key(str):Character for which frequency need to be calculated
        s (str): Input string
    Returns:
        int: Total number of vowels in the string.
  """
  fre=0
  for ch in s:
    if key==ch:fre+=1
  return fre
s=input("Enter a String")
unique=set()
for ch in s:
 if ch not in unique: 
   logging.info(f"{ch} frequency:-{fre(ch,s)}")
   unique.add(ch)
```
