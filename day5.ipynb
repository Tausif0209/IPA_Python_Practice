```python
class ten():
  def __init__(self,name,phone,email,passing_year,marks):
    self.name=name
    self.marks=marks
    self.phone=phone
    self.email=email
    self.passing_year=passing_year
  def display(self):
    print(f"10\nname:{self.name}\nmarks:-{self.marks}\nphone:{self.phone}\n")

```

```python
class twelve(ten):
  def __init__(self,name,phone,email,passing_year,marks,passing_year12,marks12):
    super().__init__(name,phone,email,passing_year,marks)
    self.passing_year12=passing_year12
    self.marks12=marks12
  def display(self):
    super().display()
    print(f"\n12\n12th passing year:-{self.passing_year12}\n12%:-{self.marks12}")
  
st=twelve("Tausif",9304340386,"tk@",2019,88,2021,89.5)
st.display()
```

```python
class bachelor(twelve):
  def __init__(self,name,phone,email,passing_year,marks,passing_year12,marks12,branch,university):
    super().__init__(name,phone,email,passing_year,marks,passing_year12,marks12)
    self.branch=branch
    self.university=university
  def display(self):
    super().display()
    print(f"\nCollege Details\nbranch:-{self.branch}\nuniversity:-{self.university}")
  
st=bachelor("Tausif",9304340386,"tk@",2019,88,2021,89.5,"IT","HIT")
st.display()
```
