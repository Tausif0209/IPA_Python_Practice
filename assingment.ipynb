```python
#Given a list of numbers, print "Valid slice" only if the first two elements are equal to the last two elements
ls = eval(input("enter list of number"))
if len(ls)>1 and ls[0]==ls[-2] and ls[1]==ls[-1]:
  print("valid slice")
```

```python
#Write a program that checks whether a string is a palindrome using slicing.
s=input("enter a string")
if s.lower()==s[::-1].lower():
  print("Palindrome")
else:
  print("Not Palindrome")
```

```python
#Given a string, check whether the last character is a digit using slicing, if digit-> Ends with digit otherwise Ends with character / Ends with special symbol.
s=input("enter a string")
if len(s)==0:print("empty string")
elif s[-1:].isdigit():
  print("End With Digit")
elif s[-1:].isalpha():
  print("End With Character")
else:
  print("End with special symbol")
```

```python
#Q4
s=input("enter a string")
if s[0].isupper():
  print("Starts with uppercase")
elif s[-1].islower():
  print("Ends with lowercase")
else:
  print("Other case")
```

```python
#Q5
ls = eval(input("enter list of number"))
size=len(ls)
mid=size//2 + size%2
sum=0
i=0
while(i<mid):
  sum+=ls[i]
  i+=1
while(i<size):
  sum-=ls[i]
  i+=1
if(sum>0):
  print("First half greater")
elif(sum==0):
  print("Equal halves")
else:
  print("Second half greater")
```
