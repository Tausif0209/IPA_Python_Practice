```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.random.randint(1,50,10)
y=np.random.randint(1,50,10)
print(x,y,sep='\n')
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,color="yellow",linestyle='dotted')
plt.show()
```

```python
plt.title("Random Chart")
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.plot(x,y,color="red",linestyle='dotted')
plt.show()
```

```python
plt.title("Random Chart")
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color="red",linestyle='-.')
plt.show()
```

```python
plt.title("Random Chart")
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color="red",linestyle='-.',alpha=0.4)#alpha values lies between [0,1] smaller the values less the intensity
plt.show()
```

```python
plt.title("Random Chart")
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color="red",linestyle='solid',alpha=0.9,marker='o')
plt.show()
```

```python
plt.title("Random Chart")
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color="red",linestyle='solid',alpha=0.4,marker='*',markerfacecolor='y')
plt.show()
```

```python
plt.title("Random Chart")
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color="red",linestyle='solid',alpha=0.4,marker='*',markerfacecolor='y',markersize=10)
plt.show()
```

```python
plt.title("Random Chart")
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color="red",linestyle='solid',alpha=0.4,marker='*',markerfacecolor='y',markeredgecolor='blue')
plt.show()
```

```python
plt.title("Random Chart")
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.xticks(x)
plt.yticks(y)
plt.plot(x,y,color="red",linestyle='solid',alpha=1,marker='*',markerfacecolor='y',markeredgecolor='blue',linewidth=6)
plt.show()
```

```python
x=np.arange(1,11)
y=np.sin(x)
z=np.cos(x)
print(x,y,z,sep="\n")
```

```python
plt.plot(x,y,label="SINQ")
plt.plot(x,z,label="COSQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.plot(x,y,label="SINQ")
plt.plot(x,z,label="COSQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.plot(x,y,label="SINQ")
plt.plot(x,z,label="COSQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")
plt.legend()


plt.subplot(1,2,2)
plt.plot(x,y,label="SINQ")
plt.plot(x,z,label="COSQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(4,4))
plt.subplot(3,3,1)
plt.plot(x,y,label="SINQ")
plt.plot(x,z,label="COSQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")
plt.legend()


plt.subplot(3,3,5)
plt.plot(x,y,label="SINQ")
plt.plot(x,z,label="COSQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")
plt.legend()

plt.subplot(3,3,9)
plt.plot(x,y,label="SINQ")
plt.plot(x,z,label="COSQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.plot(x,y,label="SINQ")
plt.plot(x,z,label="COSQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")
plt.legend()
plt.grid()


plt.subplot(1,2,2)
plt.plot(x,y,label="SINQ")
plt.plot(x,z,label="COSQ")
plt.title("SINQ & COSQ")
plt.xlabel("Data Points")
plt.ylabel("Associate Values")
plt.subplots_adjust(wspace=0.8)
plt.legend()
plt.show()
```
