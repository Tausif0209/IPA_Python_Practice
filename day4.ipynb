```python
ls=eval(input("Enter list of numbers:-"))
res=[i**2 for i in ls if i%2==0]
print(f"{ls}\n{res}")
```

```python
sen=input("Enter sentence:-")
words=[word for word in sen.split(" ") if word[0].lower() in "aeiou"]
print(f"{sen}\n{words}")
```

```python
ls=eval(input("Enter list of numbers:-"))
unique={i for i in ls if i>5}
print(f"{ls}\n{unique}")
```
