Merge and Concat

```python
data1={
  "Name":["A","B","C"],
  "age":[23,45,67],
  "salary":[1000,2000,3500]
}
```

```python
data2={
  "Name":["A","BB","C"],
  "age":[23,45,67],
  "salary":[1000,2000,3500]
}
```

```python
df1=pd.DataFrame(data1)
```

```python
df2=pd.DataFrame(data2)
```

```python
df1.head()
```

```python
df2.head()
```

```python
new_df=pd.concat([df1,df2],ignore_index=True)
```

```python
new_df.head(6)
```

```python
new_df.drop_duplicates()
```

Merge

```python
pd.merge(df1,df2,on="age",how="outer")
```
